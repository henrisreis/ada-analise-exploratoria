```python
import pandas as pd
import seaborn as sns # Biblioteca de visualização de gráficos
```

```python
import matplotlib.pyplot as plt
```

Para carregar o dataset 'Iris' da biblioteca **seaborn**:

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

Nesse dataset existe uma variável categórica importante que é a espécie da planta. É interessante tentarmos classificar a planta de acordo com as outras variáveis.

```python
iris['species'].value_counts()
```

```python
col = 'sepal_length'
sns.histplot(data=iris, x=col).set_title(f'Distribuição da variável {col}')
```

```python
iris.describe()
```

Um outro parâmetro de ser passado é o **kde = True**, que mostrará a curva da distribuição. Comparemos:

```python
sns.histplot(data=iris, x=col, kde=True).set_title(f'Distribuição da variável {col}')
```

Verifica-se que a distribuição está deslocada para a esquerda, ou seja, mais próxima de 4.3 (mínimo) do que de 7.9 (máximo).

Uma prática comum é utilizarmos um **for** para conseguirmos visualizar vários gráficos de uma vez.

```python
for col in iris:
    print(col)
```

Observa-se que dessa maneira obtemos os nomes das colunas do dataframe.

```python
for col in iris.drop(columns='species'): # É importante remover a coluna 'species' porque esse tipo de gráfico não comporta variáveis categóricas.
    sns.histplot(data=iris, x=col, kde=True).set_title(f'Distribuição da variável {col}')
    plt.show()
```

Podemos interpretar os gráficos considerando a relação que eles tem com a variável que a gente quer estudar. Para isso utilizaremos o parâmetro de cor **hue**, com a variável que queremos avaliar:

```python
for col in iris.drop(columns='species'):
    sns.histplot(data=iris, x=col, kde=True, hue=iris['species']).set_title(f'Distribuição da variável {col}')
    plt.show()
```

Dessa forma podemos avaliar como se distribuem os parâmetros dentro de cada espécie.

Utilizando um gráfico do tipo **barplot** para a análise:

```python
for col in iris.drop(columns='species'):
    sns.barplot(data=iris, x=iris['species'], hue=iris['species'], y=col, errorbar=('ci', 90))
    plt.show()
```

O parâmetro **ci** diz respeito ao **intervalo de confiança**. O valor mais utilizado geralmente é o de 90.

Avaliando os gráficos nós temos uma hipótese de que é possível fazer uma diferenciação entre as espécies com base nos parâmetros, que parecem fazer uma boa separação.
